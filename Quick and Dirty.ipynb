```python
path = 'C:/Users/kevin/Documents/Python/TCC/Credit Risk/application_test.csv'
appl_test_set = pd.read_csv(path, sep = ',')

path = 'C:/Users/kevin/Documents/Python/TCC/Credit Risk/application_train.csv'
appl_train_set = pd.read_csv(path, sep = ',')
```

```python
appl_test_set.Name = "Base de Teste"
appl_test_set.head(5)
```

```python
appl_train_set.Name = "Base de Treino"
appl_train_set.head(5)
```

```python
#Seleção das variaveis quantitativas
quant = appl_train_set.select_dtypes(['int64', 'float64'])

#Seleção das variaveis qualitativas
quali = appl_train_set.select_dtypes(['object'])

#Nome das categorias 
cat_names = quali.columns

#Parametrização das variaveis qualitativas
quali = quali.astype('category').filter(cat_names).apply(lambda x: x.cat.codes)
```

```python
#Caso algum elemento esteja nulo, ele será preenchido com o valor "-1"
df_to_fit = pd.concat([quant.fillna(-1), quali], axis= 1)
```

```python
X = df_to_fit.drop('TARGET', axis = 1)
y = df_to_fit.TARGET
```

```python
import statsmodels.api as sm
import statsmodels.discrete.discrete_model as sd
```

```python
X2 = sm.add_constant(X)
```

```python
est = sd.Logit(y, X2)
est2 = est.fit()
```

```python
qd = est2.summary()
```

```python
with open('summary.csv', 'w') as fh: 
    fh.write(qd.tables[1].as_csv())
```
