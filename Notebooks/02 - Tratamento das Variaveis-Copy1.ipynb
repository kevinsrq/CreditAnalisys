### Tratamentos das variaveis categóricas

```python
from sklearn import preprocessing
from collections import defaultdict
d = defaultdict(preprocessing.LabelEncoder)
```

```python
# Encoding the variable
fit = entrada.select_dtypes('O').apply(lambda x: d[x.name].fit_transform(x))

# Inverse the encoded
fit.apply(lambda x: d[x.name].inverse_transform(x))
```

```python
# Nomes das colunas categoricas
colunas_categoricas = entrada.select_dtypes('O').columns

# Using the dictionary to label future data
entrada[colunas_categoricas] = entrada.select_dtypes('O').apply(lambda x: d[x.name].transform(x))
```

### Aleatorização de Elementos 
--- 
Devido a capacidade computacional, é necessária a aleatorização dos elementos da base para uma boa capacidade performatica computacional. 

```python
# Tamanho da base inicialmente: 
print(f'A base contem {entrada.shape[0]} linhas e {entrada.shape[1]} colunas')
```

```python
# Elementos amostrados
amostra = entrada.sample(n=150000, replace=False, random_state=42)
```

```python
# Percentual de inadimplentes na base inicial
entrada['TARGET'].value_counts(normalize=True)
```

```python
# Percentual de inadimplentes na base amostrada
amostra['TARGET'].value_counts(normalize=True)
```

### Seleção da variaveis sem Multicolineariedade
---

```python
# Remoção das variavies que apresentam uma correlação maior que 70%
corr = amostra.corr()
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.6:
            if columns[j]:
                columns[j] = False

selected_columns = corr.columns[columns]

amostra_aux = amostra.filter(items=selected_columns)
amostra_aux.head()
```

```python
amostra_aux.rename(columns={'AMT_ANNUITY_y':'AMT_ANNUITY'}, inplace=True)
```

```python
amostra_aux.reset_index(drop=True ,)
```

## Quick and Dirty 
---

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.utils.validation import column_or_1d
```

```python
y = amostra_aux.filter(['TARGET'])
X = amostra_aux.drop(['TARGET', 'SK_ID_BUREAU', 'SK_ID_CURR'], axis=1)
```

```python
selector= SelectFromModel(estimator=LogisticRegression()).fit(X, y = column_or_1d(y, warn=True))
```

```python
selector.estimator_.coef_
```
