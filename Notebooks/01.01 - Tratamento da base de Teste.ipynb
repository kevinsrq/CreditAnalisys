### Tratamento da base Bureau
--- 

```python
biro = pd.read_csv(path+'/Bases/bureau.csv')

print(f'A base contem {biro.shape[0]} linhas e {biro.shape[1]} colunas')
biro.head(5)
```

```python
# Criação das flags para o biro (sendo esse valores apenas preenchidos para poder ser feito a regressão)
biro["DAYS_CREDIT_ENDDATE"].fillna(-1, inplace=True)
biro["DAYS_ENDDATE_FACT"].fillna(-1, inplace=True)
biro["AMT_CREDIT_MAX_OVERDUE"].fillna(-1, inplace=True)
biro["AMT_CREDIT_SUM"].fillna(-1, inplace=True)
biro["AMT_CREDIT_SUM_DEBT"].fillna(-1, inplace=True)
biro["AMT_CREDIT_SUM_LIMIT"].fillna(-1, inplace=True)
biro["AMT_ANNUITY"].fillna(-1, inplace=True)
```

### Junção das bases de Treino e Bureau
--- 

```python
enriquecida = pd.merge(apl_train_set, biro, how='left', on=['SK_ID_CURR'])

print(f'A base contem {enriquecida.shape[0]} linhas e {enriquecida.shape[1]} colunas')
enriquecida.head()
```

```python
del apl_train_set, biro
```

```python
enriquecida.query('DAYS_CREDIT >= -365', inplace=True)
enriquecida.head()
```

### Tratamento da base Bureau Balance 
--- 

```python
bureau_balance = pd.read_csv(path+'/Bases/bureau_balance.csv')

print(f'A base contem {bureau_balance.shape[0]} linhas e {bureau_balance.shape[1]} colunas')
bureau_balance.head(5)
```

```python
bureau_balance = bureau_balance.groupby(['SK_ID_BUREAU', 'STATUS']).agg({'MONTHS_BALANCE': 'min'}).reset_index()
```

### Enriquecimento com a base de balance do Biro 
---

```python
enriquecida = pd.merge(enriquecida, bureau_balance, how='left', on=['SK_ID_BUREAU'])

del bureau_balance
```

### Export da base enriquecida
--- 

```python
enriquecida.fillna(-1, inplace=True)
```

```python
enriquecida.to_csv(path+'/Processamento/base_enriquecida_teste.csv', index=False)
```
