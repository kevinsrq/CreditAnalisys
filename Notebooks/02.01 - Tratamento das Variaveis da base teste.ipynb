### Tratamentos das variaveis categÃ³ricas
--- 

```python
from sklearn import preprocessing
from collections import defaultdict
d = defaultdict(preprocessing.LabelEncoder)
```

```python
# Encoding the variable
fit = entrada.select_dtypes('O').apply(lambda x: d[x.name].fit_transform(x))

# Inverse the encoded
fit.apply(lambda x: d[x.name].inverse_transform(x))
```

```python
# Nomes das colunas categoricas
colunas_categoricas = entrada.select_dtypes('O').columns

# Using the dictionary to label future data
entrada[colunas_categoricas] = entrada.select_dtypes('O').apply(lambda x: d[x.name].transform(x))
```

### Tratamento da variavel data
---

```python
from datetime import timedelta
import datetime as dt
```

```python
# DefiniÃ§Ã£o das datas de acordo com o dias do pedido
entrada['DTREF'] = pd.to_datetime('30-08-2018', format='%d-%m-%Y')  + pd.to_timedelta(entrada['DAYS_CREDIT'], unit='days')

# VisualizaÃ§Ã§Ã£o dos primeiros elementos das datas
entrada['DTREF'].head(4)
```

```python
# DefiniÃ§Ã£o das safras
entrada['SAFRA'] = entrada['DTREF'].dt.strftime('%Y%m')
```
